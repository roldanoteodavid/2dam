Dibuja las siguientes funciones:
$$ y = A*x $$
$$ y = A*x^2 $$

* (0.5 puntos) La fuente de título es : "Arial" o "DejaVu Sans", de color darkred y tamaño 16.
* (0.5 puntos) Los colores de las gráficas son verde, rojo y azul, la linea de 1 y el grid que aparezca por defecto
* (0.5 puntos) Dibuja una linea vertical que vaya hasta la solución y leyenda con ecuaciones.
* (0.5 puntos) Marca la solución con una $\star$ y pon las etiquetas en cada eje


<img src="matplotlib.png" ALIGN="left" width=200 height=200)>


````python
import numpy as np
import matplotlib.pyplot as plt

A = 15

# Definir las ecuaciones
def ecuacion1(y):
    return A * y

def ecuacion2(y):
    return A * (y*y)

# Crear valores de x
y = np.linspace(0.0, 1.4, 100)

# Apartado 1
font = {'family': 'Arial', 'color': 'darkred', 'size': 16}
plt.title("Sistema de ecuaciones", loc='center', fontdict=font)
plt.xlabel("x")
plt.ylabel("y")

# Apartado 2
plt.plot(y, ecuacion1(y), label=r'$y=A*x$', color='g', linewidth=1)
plt.plot(y, ecuacion2(y), label=r'$y=A*x^2$', color='r', linewidth=1)

# Apartado 4
sol_x = 1
sol_y = 15
plt.scatter(sol_x, sol_y, color='b', marker='*', label='Solución')

# Apartado 3
plt.axvline(x=sol_x, ymax= 0.5, color='b', linestyle='-', linewidth=1)

# Mostrar la leyenda y la cuadrícula
plt.legend()
plt.grid()

# Mostrar el gráfico
plt.show()

````

````python
#Descomenta el siguiente código para probar la clase:

class Repartidor:
    def __init__(self, nombre="Nombre", zona_entrega = "Zona", pedidos_entregados = 0): 
        self.nombre = nombre
        self. zona_entrega =  zona_entrega
        self.pedidos_entregados = pedidos_entregados
        
    
    def entregar_pedido(self):
        self.pedidos_entregados = self.pedidos_entregados + 1
        print(self.nombre + " ha entregado un pedido en la zona " + self.zona_entrega)
    
    def informe_entregas(self):
        print(self.nombre + " ha entregado " + str(self.pedidos_entregados) + " pedidos en total.")


repartidor1 = Repartidor("Carlos", "Centro")

repartidor1.entregar_pedido()
repartidor1.entregar_pedido()
repartidor1.entregar_pedido()

repartidor1.informe_entregas()
````

Una tragaperras con 3 rodillos y 6 símbolos diferentes en cada rodillo tiene los siguiente premios:

* 3 símbolos iguales: 5 monedas.
* 2 símbolos iguales: 1 monedas.
* ningún símbolo igual: -1 moneda.

Escribe un programa que simule el juego y calcule las ganancias o perdidas totales después de un número 1000 jugadas si empezamos con 100 monedas. 

````python
import random

num_tiradas = 1000

def girartragaperras():
    
    simbolos = [random.randint(1, 6),random.randint(1, 6), random.randint(1, 6)]
    
    if simbolos[0] == simbolos[1] and simbolos[1] == simbolos[2]:
        balance = 5
    elif (simbolos[0] == simbolos[1] and simbolos[1] != simbolos[2]) or (simbolos[0] != simbolos[1] and simbolos[1] == simbolos[2]) or (simbolos[0] != simbolos[1] and simbolos[0] == simbolos[2]):
        balance = 1
    elif  simbolos[0] != simbolos[1] and simbolos[1] != simbolos[2] and simbolos[0] != simbolos[2]:
        balance = -1
    
    return balance

def simulacion():
    
    dinero = 100

    for i in range(1,num_tiradas):
        balance = girartragaperras()
        dinero = dinero + balance
    
    dineroFinal = dinero
    
    return dineroFinal

ganancias = simulacion() - 100

if ganancias < 0:
    print(f"Después de {num_tiradas} tiradas, las perdidas totales son: {ganancias} monedas.")
elif ganancias > 0:
    print(f"Después de {num_tiradas} tiradas, las ganancias totales son: {ganancias} monedas.")
else:
    print(f"Después de {num_tiradas} tiradas, sigues con las mismas monedas.")
````
