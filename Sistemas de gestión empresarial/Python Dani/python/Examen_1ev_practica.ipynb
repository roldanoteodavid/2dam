Dibuja el siguiente sistema de ecuaciones:
$$ x+2y=3 $$
$$ x+y=2 $$

* Marca la solución con una $\star$
* Utiliza para las gráficas un ancho del linea de 1.
* Los colores son verde, rojo y azul. 
* Dibuja una linea vertical que vaya hasta la solución.
* La fuente de título es : "Arial" o "DejaVu Sans", de color darkred y tamaño 16

<img src="matplotlib.png" ALIGN="left" width=200 height=200)>


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,2.5,1)
y = (3-x)/2
y2 = 2-x
font= {'family':'Arial','color':'darkred','size':16}
plt.title("Ejercicio 1", loc = 'center', fontdict = font)
plt.plot(x,y,linewidth = '1',color='g', label=r'$y=\frac{-x+3}{2}$')
plt.plot(x,y2, linewidth = '1',color='r', label=r'$y={2-x}$' )
plt.xlabel("x")
plt.ylabel("y")
plt.axvline(x=1, ymax=0.5,color='b', marker = '*', linewidth = '1', label=r'Solución')
plt.legend()
plt.grid()
plt.show()
```

<b><font color='red'>Ejercicio 02 (2.5 puntos)</font></b> 

Tira tres monedas, realiza el suficiente número de tiradas para responder la siguiente pregunta.
¿Cuál sería  la probabilidad de obtener tres cruces?

```python
import random

def simular_tiradas(n):
    eventos_exitosos = 0

    for _ in range(n):
        # Tirar tres monedas (cara = 0, cruz = 1)
        resultado_tirada = [random.randint(0, 1) for _ in range(3)]

        # Verificar si todas son cruces
        if all( resultado == 1 for resultado in resultado_tirada):
            eventos_exitosos += 1

    probabilidad = eventos_exitosos / n
    return probabilidad

# Realizar la simulación con un número suficiente de tiradas
numero_tiradas = 1000000
probabilidad_obtener_tres_cruces = simular_tiradas(numero_tiradas)

print(f"Probabilidad de obtener tres cruces: {probabilidad_obtener_tres_cruces:.4f}")

print("Probabilidad de obtener tres cruces = " + str(simular_tiradas(numero_tiradas)))
```

Crea una función llamada pintar_tablero, que reciba una lista con las fichas y las saque por pantalla


```python
import numpy as np

def imprimir_tablero(posfichas):
    tablero = np.full((8, 8), ' ', dtype=str)

    for ficha in posfichas:
        fila = 8 - int(ficha.pos[1])
        columna = ord(ficha.pos[0]) - ord('a')
        tablero[fila, columna] = ficha.caracter

    for fila in tablero:
        print(list(fila))

# Crear objetos Ficha y agregarlos a posfichas
#posfichas = [Ficha(i) for i in inicio]

# Llamar a la función para imprimir el tablero
imprimir_tablero(posfichas)
```
